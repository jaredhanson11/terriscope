```python
import json
import requests
```

```python
# Load the payload from import_payload.json
with open('import_payload.json', 'r') as f:
    payload = json.load(f)

# Define the API endpoint
url = 'http://localhost:8000/maps'

# Make the POST request
response = requests.post(url, json=payload)

# Print the response
print(f"Status Code: {response.status_code}")
```
